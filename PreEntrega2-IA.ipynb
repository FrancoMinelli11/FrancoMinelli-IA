```python
import openai

# Configuración de la API de OpenAI
openai.api_key = 'XXXXXXXXXXXXXXXXX'

# Función para generar código con GPT-4 utilizando el endpoint de chat
def generar_codigo_chat(messages, max_tokens=150, temperature=0.5):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        max_tokens=max_tokens,
        temperature=temperature,
    )
    return response.choices[0].message['content'].strip()

# Función para generar imágenes con DALL-E
def generar_imagen(prompt, n=1, size="1024x1024"):
    response = openai.Image.create(
        prompt=prompt,
        n=n,
        size=size
    )
    return response['data'][0]['url']

# Mensajes para el modelo de chat con few-shot prompting
messages = [
    {"role": "system", "content": "Eres un experto desarrollador web."},
    {"role": "user", "content": "Escribe un código HTML básico para una página web que incluya: un título, un encabezado principal y un párrafo de bienvenida."},
    # Ejemplo 1
    {"role": "assistant", "content": """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Mi Primer Sitio Web</title>
</head>
<body>
    <h1>Bienvenido</h1>
    <p>Este es mi primer sitio web. Espero que disfrutes la visita.</p>
</body>
</html>
    """},
    # Nueva solicitud para el modelo
    {"role": "user", "content": "Escribe un código HTML básico para una página web que incluya: un título, un encabezado principal y un párrafo de bienvenida. El título debe ser 'Bienvenido a mi sitio web', el encabezado principal debe ser 'Hola Mundo' y el párrafo debe decir 'Gracias por visitar mi sitio web. Aquí encontrarás información interesante.'"}
]

# Generar código
codigo_generado_chat = generar_codigo_chat(messages)
print("Código HTML generado:")
print(codigo_generado_chat)

# Prompt para DALL-E
prompt_imagen = "A beautiful website homepage design with a welcoming theme"

# Generar imagen
url_imagen = generar_imagen(prompt_imagen)
print("URL de la imagen generada:")
print(url_imagen)
```
